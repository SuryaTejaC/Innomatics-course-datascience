```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
# initializing the lists

clean_quora_lst = []

quora_len_lst = []
```

```python
def preprocess(raw_tweet, flag):
    # Removing special characters and digits
    letters_only = re.sub("[^a-zA-Z]", " ",raw_tweet)
    
    # change sentence to lower case
    letters_only = letters_only.lower()

    # tokenize into words
    words = letters_only.split()
    
    # remove stop words                
    words = [w for w in words if not w in stopwords.words("english")]
    
    #Stemming/Lemmatization
    if(flag == 'stem'):
        words = [stemmer.stem(word) for word in words]
    else:
        words = [lemmatizer.lemmatize(word) for word in words]
    
    clean_quora_lst.append(" ".join(words))
    
    quora_len_lst.append(len(words))
```

```python
# lemmatization from question1
clean_quora_lst = []

quora_len_lst = []
sample_df['question1'].apply(lambda x: preprocess(x, 'lemma'))

sample_df['clean_quora_lemma_q1'] = clean_quora_lst

sample_df['quora_length_lemma_q1'] = quora_len_lst
```

```python
clean_quora_lst = []

quora_len_lst = []
sample_df['question2'].apply(lambda x: preprocess(x, 'lemma'))

sample_df['clean_quora_lemma_q2'] = clean_quora_lst

sample_df['quora_length_lemma_q2'] = quora_len_lst
```

```python
clean_quora_q1=[]
for tweet in sample_df['clean_quora_lemma_q1']:
    clean_quora_q1.append(tweet)

clean_quora_q2=[]
for tweet in sample_df['clean_quora_lemma_q2']:
    clean_quora_q2.append(tweet)
```
